```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def predict_poly(x, koeff):
    res = 0
    xx = [x ** (len(koeff) - n - 1) for n in range(len(koeff))]

    for i, k in enumerate(koeff):
        res += k * xx[i]

    return res
```

```python
x = np.arange(0, 10.1, 0.1)
y = 1 / (1 + 10 * np.square(x))

x_train, y_train = x[::2], y[::2]

N = len(x)
M = 54

z_train = np.polyfit(x_train, y_train, M)
print(z_train)

a = z_train
xplt = y
# yplt = [a[0]*xx**2 + a[1]*xx + a[2] for xx in x]
yplt = np.zeros(N)
for i in range(N):
  ytmp = 0.0
  for j in range(M + 1):
    ytmp += a[j] * x[i] ** (M - j)
  yplt[i] = ytmp
# print(yplt)
plt.figure(figsize=(12, 10))
plt.scatter([j for j in range(len(x))], xplt, color='blue', linestyle="-.", marker='o', label='01')
plt.scatter([j for j in range(len(x))], yplt, color='red', linestyle="-.", marker='x', label='01')

# plt.xlim([0, 45])
# plt.ylim([0, 1])
plt.ylabel("длина")
plt.xlabel("ширина")
plt.grid(True)
plt.show()
```
